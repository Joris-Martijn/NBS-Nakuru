```python
#Q_calibrated_path = os.path.join(folder, "output_fx0.4_ts_x8_kh_x0.05_kvx3_L5_2007_2021.csv")
Q_calibrated_path = os.path.join(folder, "output.csv")
Q_calibrated = pd.read_csv(Q_calibrated_path, sep=",", header=0, index_col=0, parse_dates=True)
display(Q_calibrated)
gauges = ['Q_2', 'Q_3', 'Q_4', 'Q_5', 'Q_6', 'Q_7', 'Q_8', 'Q_14', 'Q_17']
```

```python
# Assume Q_calibrated is your DataFrame with datetime index and gauge columns
# gauges = ['Q_2', 'Q_3', ...]  # already defined
return_periods = [2, 5, 10]
probabilities = [1 - 1/t for t in return_periods]

results = []

for gauge in gauges:
    # Filter vanaf 1 januari 2007
    Q_series = Q_calibrated[gauge]
    # Get annual maxima
    annual_maxima = Q_series.resample('Y').max().dropna()
    # Fit Gumbel distribution
    params = stats.gumbel_r.fit(annual_maxima)
    # Calculate quantiles for each return period
    discharge_return_levels = stats.gumbel_r.ppf(probabilities, *params)
    # Store results
    for T, Q in zip(return_periods, discharge_return_levels):
        results.append({'Gauge': gauge, 'ReturnPeriod': T, 'Discharge_m3s': Q})

# Convert to DataFrame for easy viewing
df_return_periods = pd.DataFrame(results)


# Adjust Q_17 and Q_2 for all return periods
df_return_periods_adj = df_return_periods.copy()

for T in return_periods:
    # Get initial (unadjusted) values
    Q_17_init = df_return_periods.loc[
        (df_return_periods['Gauge'] == 'Q_17') & (df_return_periods['ReturnPeriod'] == T),
        'Discharge_m3s'
    ].values[0]
    Q_3_init = df_return_periods.loc[
        (df_return_periods['Gauge'] == 'Q_3') & (df_return_periods['ReturnPeriod'] == T),
        'Discharge_m3s'
    ].values[0]
    Q_2_init = df_return_periods.loc[
        (df_return_periods['Gauge'] == 'Q_2') & (df_return_periods['ReturnPeriod'] == T),
        'Discharge_m3s'
    ].values[0]

    # Q_17 = initial Q_17 - initial Q_3
    df_return_periods_adj.loc[
        (df_return_periods_adj['Gauge'] == 'Q_17') & (df_return_periods_adj['ReturnPeriod'] == T),
        'Discharge_m3s'
    ] = Q_17_init - Q_3_init

    # Q_2 = initial Q_2 - initial Q_17
    df_return_periods_adj.loc[
        (df_return_periods_adj['Gauge'] == 'Q_2') & (df_return_periods_adj['ReturnPeriod'] == T),
        'Discharge_m3s'
    ] = Q_2_init - Q_17_init

# Use df_return_periods_adj for further processing
df_return_periods = df_return_periods_adj

display(df_return_periods)
#df_return_periods.to_csv(os.path.join(output_folder, "Q_calibrated_return_periods_terracing_all_subcatchments.csv"), sep=",", index=False)

# Filter for gauges Q_9 and Q_10
#df_gauges_9_10 = df_return_periods[df_return_periods['Gauge'].isin(['Q_9', 'Q_10'])]

# Save to Excel
#df_gauges_9_10.to_excel(os.path.join(folder, "return_periods_Q9_Q10_with_cc_ext.xlsx"), index=False)
```

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

# SCS-like hourly fraction distribution (sums to 1.00)
hour_fractions_flattened = np.array([0.03, 0.05, 0.07, 0.09, 0.10, 0.11, 0.12, 0.13,
                                     0.13, 0.12, 0.11, 0.10, 0.09, 0.07, 0.06, 0.05,
                                     0.04, 0.03, 0.02, 0.01, 0.00, 0.00, 0.00, 0.00])
hour_fractions_flattened /= hour_fractions_flattened.sum()

steps_per_hour = 12  # 5-min steps
total_steps = 24 * steps_per_hour
hours = np.arange(0, 24, 1/steps_per_hour)  # 0.00, 0.083..., ..., 23.916...

interp_hours = np.arange(24)

# Make sure delay DataFrame is indexed by 'Gauge'
delay = delay.set_index('Gauge')

return_periods = [2, 5, 10]  # Return periods in years

for T in return_periods:
    fig, axs = plt.subplots(3, 3, figsize=(12, 8), sharex=True)
    fig.suptitle(f'5-min Discharge Curves per Gauge (T={T} yr)', fontsize=16, y=1.0)
    axs = axs.flatten()
    for i, gauge in enumerate(gauges):
        # Get the discharge for this gauge and return period
        Q = df_return_periods.loc[
            (df_return_periods['Gauge'] == gauge) & (df_return_periods['ReturnPeriod'] == T),
            'Discharge_m3s'
        ].values[0]
        
        daily_volume = Q * 86400  # in mÂ³
        hourly_discharge = (hour_fractions_flattened * daily_volume) / 3600  # in mÂ³/s

        # Interpolate to 5-min steps
        interp_func = np.interp(hours, interp_hours, hourly_discharge)

        # Get travel time (in hours) for this gauge, default to 0 if not found
        travel_time_hr = float(delay.loc[gauge, 'Travel time (hours)']) if gauge in delay.index else 0
        travel_steps = int(round(travel_time_hr * steps_per_hour))  # shift in 5-min steps

        # Shift the discharge curve on 5-min resolution
        interp_func_shifted = np.roll(interp_func, travel_steps)

        # Format time as HH:MM
        time_labels = [f"{int(h):02d}:{int((h%1)*60):02d}" for h in hours]

        # Save to DataFrame
        df_5min = pd.DataFrame({
            'Time': time_labels,
            'Discharge': interp_func_shifted
        })

        # Save as CSV and .dat (no header, utf-8)
        csv_path = os.path.join(output_folder, f"{gauge}_T{T}_5min_discharge_SWMM_terracing_all_subcatchments.csv")
        dat_path = os.path.join(output_folder, f"{gauge}_T{T}_5min_discharge_SWMM_terracing_all_subcatchments.dat")
        df_5min.to_csv(csv_path, index=False, header=False, encoding='utf-8', sep="\t")
        df_5min.to_csv(dat_path, index=False, header=False, encoding='utf-8', sep="\t")

        # Plot
        axs[i].plot(time_labels, interp_func_shifted, color='steelblue')
        axs[i].set_title(f'{gauge}')
        axs[i].set_xlabel('Time (HH:MM)', fontsize=12)
        axs[i].set_ylabel('Discharge (mÂ³/s)', fontsize=12)
        axs[i].xaxis.set_major_locator(plt.MaxNLocator(6))
        axs[i].grid(True)

    plt.tight_layout(rect=[0,0,1,0.99])  # Adjust layout to fit suptitle	
    plt.savefig(os.path.join(output_folder, f"5min_discharge_curves_T{T}_terracing_all_subcatchments.png"), dpi=300)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# SCS-like hourly fraction distribution (sums to 1.00)
hour_fractions_flattened = np.array([
    0.03, 0.05, 0.07, 0.09, 0.10, 0.11, 0.12, 0.13,
    0.13, 0.12, 0.11, 0.10, 0.09, 0.07, 0.06, 0.05,
    0.04, 0.03, 0.02, 0.01, 0.00, 0.00, 0.00, 0.00
])
hour_fractions_flattened /= hour_fractions_flattened.sum()

hours = np.arange(24)
normalized_curve = hour_fractions_flattened / hour_fractions_flattened.max()

delay_hours = 3  # Example delay

curve_hours = np.arange(delay_hours + len(normalized_curve))

plt.figure(figsize=(9, 4))
# Plot the delay part as a horizontal dotted line at zero
plt.plot(curve_hours[:delay_hours], np.zeros(delay_hours), 'k:', label='Delay', linewidth=2, linestyle=':')
# Plot the discharge curve after the delay as a solid line, starting exactly at the end of the dotted line
plt.plot(curve_hours[delay_hours-1:], np.concatenate(([0], normalized_curve)), 'b-', label='Normalized Discharge', linewidth=2)

plt.xlabel('Hour', fontsize=12)
plt.ylabel('Normalized Discharge', fontsize=12)
plt.title('Normalized Discharge Curve with Delay', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join("Figures/normalized_discharge_curve_with_delay.png"), dpi=300)
plt.show()
```
